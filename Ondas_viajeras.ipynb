La función

$f(x')=e^{-\frac{(x'-0.5)^2}{0.01}}$

Representa una perturbación en una cuerda

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-3,4,400)

f=np.exp(-(x-0.5)**2 /0.01)

plt.plot(x,f)
plt.show()
```

Esta es la gráfica original $f(x)$

```python
import numpy as np
import matplotlib.pyplot as plt

c=1
t=1
x1=x-c*t
x=np.linspace(-3,4,400)

f=np.exp(-(x1-0.5)**2 /0.01)

plt.plot(x,f)
plt.title('Desplazamiento a la Derecha')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

Aquí hicimos el cambio de variable $x'=x-ct$ tomando los valores de $ c $ y de $ t $ como constantes que pueden modificarse
Se aprecia que la gráfica se desplaza a la derecha

```python
import numpy as np
import matplotlib.pyplot as plt

c=1
t=1
x1=x+c*t
x=np.linspace(-3,4,400)

f=np.exp(-(x1-0.5)**2 /0.01)

plt.plot(x,f)
plt.title('Desplazamiento a la Izquierda')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

Aquí hicimos el cambio de variable $x'=x+ct$ tomando los valores de $ c $ y de $ t $ como constantes que pueden modificarse
Nótese que la gráfica se mueve hacia la izquierda

```python
import numpy as np
import matplotlib.pyplot as plt

c = 1
x = np.linspace(-1, 2, 400)

for t in np.arange(0, 1, 0.2):
    x1 = x + c * t
    x2 = x - c * t

    f=np.exp(-(x-0.5)**2 /0.01)
    f_l = np.exp(-(x1 - 0.5)**2 / 0.01)
    f_r = np.exp(-(x2 - 0.5)**2 / 0.01)

    plt.plot(x, f, label='Función original')
    plt.plot(x, f_l, label='Desplazamiento a la izquierda')
    plt.plot(x, f_r, label='Desplazamiento a la derecha', linestyle='--')
    plt.title('Superposición de ondas viajeras en t = ' + str(t))
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

Aquí se ve la superposición de las gráficas para diferentes tiempos, notemos que la gráfica azul se mueve a la izquierda y la amarilla a la derecha

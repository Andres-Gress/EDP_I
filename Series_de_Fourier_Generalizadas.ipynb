PROBLEMA DE STURM-LIOUVILLE

$f''(x)+2f'(x)+\lambda f(x)=0$ $ \quad $ $ 0<x<1 $

$f(0)=0, \quad f(1)=0$


Notemos que es un problema de Sturm, pero necesitamos agregar un factor integrante para verlo claramente, en este caso el factor integrante es $e^{2x}$

De modo que con esto se aprecia que

$p(x)=e^{2x}, \quad \sigma(x)=e^{2x}, \quad g(x)=0, \quad a=0, \quad b=1, \quad k_1=k_2=1, \quad k_2=k_4=0$

La solución del problema se Sturm es

$f(x)=e^{-x}sen(n\pi x) $ con $\lambda=1-(n\pi)^2$

```python
import sympy as sp
sp.init_printing()

x = sp.symbols('x')

u = sp.exp(-x)

display(u)

sp.plot(u, (x, 0, 1), title='Gráfica de $e^{-x}$', ylabel='y')
```

Gráfica de la función $u=e^{-x}$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x', real=True)
pi = sp.pi

u = sp.exp(-x)
h= sp.exp(2*x)

sumatoria = 0

for n in range(1, 100):
    k = 1 - (n * pi)**2
    omega = sp.sqrt(1 - k)
    f = sp.exp(-x) * sp.sin(omega * x)

    integrando_1 = u * f*h
    integral_1 = sp.integrate(integrando_1, (x, 0, 1))

    f_squared = f**2*h
    integral_2 = sp.integrate(f_squared, (x, 0, 1))

    division = integral_1 / integral_2

    sumatoria += division * f

sumatoria_func = sp.lambdify(x, sumatoria, 'numpy')

x_vals = np.linspace(0, 1, 400)

y_vals_sumatoria = sumatoria_func(x_vals)
y_vals_exp = np.exp(-x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals_sumatoria, label='Sumatoria S(x)', color='b')
plt.plot(x_vals, y_vals_exp, label='$e^{-x}$', color='r', linestyle='--')
plt.title('Gráfica de la Sumatoria S(x) y $e^{-x}$ para n=1 a 100')
plt.xlabel('x')
plt.ylabel('Valor')
plt.grid(True)
plt.legend()
plt.show()
```

Serie generalizada de Fourier junto con la función que se deseaba aproximar, en este caso para n=100

<font color="red">**SERIE DE FOURIER DE**</font> $\quad f(x)= \left\{\begin{matrix}
0 \quad \text{si} \quad -2\leq x<0 \\ 2-x \quad si \quad 0\leq x \leq 2
\end{matrix}\right.$

```python
import numpy as np
import matplotlib.pyplot as plt
```

Aquí podemos observar la importación de las librerias numpy y matploit para graficar

```python
def u(x,N=200):
  f=np.zeros(len(x))
  g=np.zeros(len(x))
  for n in range(1,N):
    g+=(2*(1+(-1)**(n+1))/(n*np.pi)**2)*np.cos(n*np.pi*x/2)+(2/(n*np.pi))*np.sin(n*np.pi*x/2)
  f=1/2+g
  return f
```

Definimos la función de la serie de fourier

```python
x1=np.linspace(-2,0,200)
x2=np.linspace(0,2,200)
x3=np.linspace(-2,2,200)
```

Definimos los limites de $x$ para la graficación de 3 graficas, dos que pertenecen a una función a trozos y la de la serie de fourier, por lo que ocupare 3 intervalos diferentes para cada función

```python
y1=np.zeros(len(x1))
y2=2-x2
y3=u(x3,N=200)
```

Aquí definimos el valor de cada función, la primera es cuando $x$ se encuentra en el intervalo $[0,2]$, la segunda cuando $x$ va de $[0,2]$ y la serie que va de $[-2,2]$ cos sus respectivas x indicadas anteriormente definiendo su intervalo

```python
plt.title("Variación de Armónicos")

plt.plot(x1,y1,label='y=0')
plt.plot(x2,y2,label='y=2-x')
plt.plot(x3,y3,label='u(N=200)')

for i in range(5,15,3):
    plt.plot(x3,u(x3,i),label='$f_N(x)$ con $N=%d$' %i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$');


plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

Finalmente grafico las funciones en donde vareo la cantidad de armonicos con un ciclo for para apreciar que la función donde no está definida pasa por el promedio, es decir $(0,1)$ y entre mayor número de armonicos, más se parece a la función definida a trozos
